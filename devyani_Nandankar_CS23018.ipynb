```python
# Generate numbers from 0 to 1000
numbers = list(range(0, 1001))

# Separate odd and even numbers
ODD_LIST = []
EVEN_LIST = []

for num in numbers:
    if num % 2 == 0:
        EVEN_LIST.append(num)
    else:
        ODD_LIST.append(num)

# Reverse both lists
ODD_LIST.reverse()
EVEN_LIST.reverse()

# Convert lists to tuples
ODD_TUPLE = tuple(ODD_LIST)
EVEN_TUPLE = tuple(EVEN_LIST)

# Print results
print("Reversed ODD_LIST as tuple:")
print(ODD_TUPLE)

print("\nReversed EVEN_LIST as tuple:")
print(EVEN_TUPLE)

```

```python
import numpy as np

# Input for 5x3 matrix
print("Enter elements for 5x3 matrix:")
a = []
for i in range(5):
    row = input(f"Row {i+1}: ").split()
    row = [float(x) for x in row]
    a.append(row)

# Input for 3x2 matrix
print("\nEnter elements for 3x2 matrix:")
b = []
for i in range(3):
    row = input(f"Row {i+1}: ").split()
    row = [float(x) for x in row]
    b.append(row)

# Convert to numpy arrays
A = np.array(a)
B = np.array(b)

# Multiply matrices
product = np.dot(A, B)

# Print result
print("\nProduct matrix (5x2):")
print(product)
```

```python
def is_palindrome(s):
    s = s.replace(" ", "").lower()
    return s == s[::-1]

print(is_palindrome("madam"))       # True
print(is_palindrome("nurses run"))  # True
print(is_palindrome("hello"))       # False
```

```python
def is_perfect(n):
    return n > 0 and sum(i for i in range(1, n) if n % i == 0) == n

print(is_perfect(6))   # True
print(is_perfect(28))  # True
print(is_perfect(12))  # False
```

```python
#Devyani Nandankar(CS23018)
string = input("Enter a string: ")
char_count = {}
for char in string:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1
max_char = None
max_count = 0
for char in char_count:
    if char_count[char] > max_count:
        max_count = char_count[char]
        max_char = char
print(f"Character with maximum frequency: '{max_char}' ({max_count} times)")
```

```python
#Devyani Nandankar(CS23018)
def calculate_painting_cost():
    interior_cost = 18
    exterior_cost = 12

    num_interior_walls = int(input("Enter the number of Interior walls: "))
    num_exterior_walls = int(input("Enter the number of Exterior walls: "))

    total_cost = 0

    for i in range(num_interior_walls):
        area = float(input(f"Enter the surface area of Interior wall {i+1}: "))
        total_cost += area * interior_cost

    for i in range(num_exterior_walls):
        area = float(input(f"Enter the surface area of Exterior wall {i+1}: "))
        total_cost += area * exterior_cost

    print(f"Total cost of painting: Rs. {total_cost:.2f}")

calculate_painting_cost()
```

```python
#Devyani Nandankar(CS23018)
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32


celsius = float(input("Enter the temperature in Celsius: "))


fahrenheit = celsius_to_fahrenheit(celsius)


print(f"Temperature: {celsius}°C = {fahrenheit}°F")


if celsius < -273.15:
    print("The temperature is invalid because it is below absolute zero.")
elif celsius == -273.15:
    print("The temperature is absolute 0.")
elif -273.15 < celsius < 0:
    print("The temperature is below freezing.")
elif celsius == 0:
    print("The temperature is at the freezing point.")
elif 0 < celsius < 100:
    print("The temperature is in the normal range.")
elif celsius == 100:
    print("The temperature is at the boiling point.")
else:
    print("The temperature is above the boiling point.")
```

```python
#Devyani Nandankar(CS23018)
def is_leap_year(year):

    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False


year = int(input("Enter a year: "))


if is_leap_year(year):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```

```python
#Devyani Nandankar(CS23018)
def calculate_bonus(percentage, salary):

    if 90 <= percentage <= 100:
        bonus = salary * 0.20
    elif 80 <= percentage < 90:
        bonus = salary * 0.10
    elif 70 <= percentage < 80:
        bonus = salary * 0.05
    else:
        bonus = 0
    return bonus


salary = float(input("Enter the employee's salary: "))
percentage = float(input("Enter the employee's performance percentage: "))
bonus = calculate_bonus(percentage, salary)

if bonus > 0:
    print(f"The employee's bonus is: Rs.{bonus}")
else:
    print("The employee is not eligible for a bonus.")
```

```python
#Devyani Nandankar(CS23018)
def check_loan_eligibility(credit_score):
    if credit_score >= 750:
        return "Eligible for all loan types."
    elif 650 <= credit_score < 750:
        return "Eligible for most loan types."
    elif 550 <= credit_score < 650:
        return "Eligible for some loan types with higher interest rates."
    else:
        return "May not be eligible for a loan."

credit_score = int(input("Enter the applicant's credit score: "))

eligibility = check_loan_eligibility(credit_score)

print(eligibility)
```
